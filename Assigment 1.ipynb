```python
#To visualize the distribution and relationship of numerical and categorical variables in the dataset.
#Histogram for numerical variables
import pandas as pd
df = pd.read_csv("C:\\Users\\moham\\Desktop\\Assigments\\Basic stats - 1\\sales_data_with_discounts.csv")
df
```

```python
# Plot histograms for each numerical column.
# Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences

import matplotlib.pyplot as plt

for col in numerical_cols:
  plt.figure(figsize=(8, 6))
  plt.hist(df[col], bins=20)
  plt.xlabel(col)
  plt.ylabel("Frequency")
  plt.title(f"Histogram of {col}")
  plt.show()

  # Analyze distribution and provide inferences
  print(f"\nAnalysis of {col}:")
  if df[col].skew() > 0:
    print("The distribution is positively skewed (right-skewed).")
    print("This might indicate the presence of outliers with high values.")
  elif df[col].skew() < 0:
    print("The distribution is negatively skewed (left-skewed).")
    print("This might indicate the presence of outliers with low values.")
  else:
    print("The distribution is approximately symmetrical.")

  # Basic outlier detection (you can refine this based on your specific needs)
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  if not outliers.empty:
    print("Potential outliers detected:")
    print(outliers)
  else:
    print("No significant outliers detected based on IQR.")
```

```python
#BOXPLOT
#Create boxplots for numerical variables to identify outliers  and the interquartile range.Discuss any findings, such as extreme values or unusual distributions.

import matplotlib.pyplot as plt
import seaborn as sns

# Create boxplots for numerical variables
for col in numerical_cols:
  plt.figure(figsize=(8, 6))
  sns.boxplot(x=df[col])
  plt.xlabel(col)
  plt.title(f"Boxplot of {col}")
  plt.show()

  # Analyze the boxplot and discuss findings
  print(f"\nAnalysis of {col} Boxplot:")
  # Check for outliers
  if len(df[df[col] > df[col].quantile(0.75) + 1.5 * (df[col].quantile(0.75) - df[col].quantile(0.25))]) > 0:
    print("Outliers detected above the upper quartile.")
  if len(df[df[col] < df[col].quantile(0.25) - 1.5 * (df[col].quantile(0.75) - df[col].quantile(0.25))]) > 0:
    print("Outliers detected below the lower quartile.")
  # Check for skewness
  if df[col].skew() > 0.5:
    print("The distribution is positively skewed (right-skewed).")
  elif df[col].skew() < -0.5:
    print("The distribution is negatively skewed (left-skewed).")
  else:
    print("The distribution is relatively symmetrical.")

  # Discuss the interquartile range (IQR)
  iqr = df[col].quantile(0.75) - df[col].quantile(0.25)
  print(f"The interquartile range (IQR) is {iqr}.")
  print("The IQR represents the spread of the middle 50% of the data.")
```

```python
#   Explain the concept of standardization (z-score normalization).
# 	Standardize the numerical columns using the formula: z=x-mu/sigma
# 	Show before and after comparisons of the data distributions.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Assuming df and numerical_cols are already defined from the preceding code

# Explanation of Standardization (Z-score Normalization)
print("## Standardization (Z-score Normalization)")
print(
    "Standardization, also known as z-score normalization, is a data preprocessing technique that transforms the data to have a mean of 0 and a standard deviation of 1. This is achieved by subtracting the mean of the feature from each data point and then dividing by the standard deviation."
)
print(
    "The formula for standardization is: z = (x - mu) / sigma, where:"
)
print("- z is the standardized value.")
print("- x is the original value.")
print("- mu is the mean of the feature.")
print("- sigma is the standard deviation of the feature.")
print(
    "Standardization is beneficial for machine learning algorithms that are sensitive to feature scaling, such as those that use distance calculations or gradient descent."
)

# Standardize the numerical columns using the formula z = (x - mu) / sigma
for col in numerical_cols:
  mu = df[col].mean()
  sigma = df[col].std()
  df[col + "_standardized"] = (df[col] - mu) / sigma

# Show before and after comparisons of the data distributions
for col in numerical_cols:
  # Before standardization
  plt.figure(figsize=(12, 4))
  plt.subplot(1, 2, 1)
  sns.histplot(df[col], kde=True)
  plt.title(f"Distribution of {col} (Before Standardization)")

  # After standardization
  plt.subplot(1, 2, 2)
  sns.histplot(df[col + "_standardized"], kde=True)
  plt.title(f"Distribution of {col} (After Standardization)")
  plt.show()

print(
    "As you can see from the histograms, after standardization, the data is centered around 0 with a standard deviation of 1."
)
```

```python
#Conversion of Categorical Data into Dummy Variables
# To transform categorical variables into a format that can be provided to ML algorithms.

import pandas as pd
# Import the necessary library
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder object
encoder = OneHotEncoder(handle_unknown='ignore')

# Fit and transform the categorical columns
encoded_data = encoder.fit_transform(df[categorical_cols])

# Convert the sparse matrix to a dense array
encoded_array = encoded_data.toarray()

# Create a new DataFrame with the encoded features
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_cols))

# Concatenate the encoded DataFrame with the original DataFrame
df_encoded = pd.concat([df, encoded_df], axis=1)

# Remove the original categorical columns if you don't need them anymore
df_encoded = df_encoded.drop(categorical_cols, axis=1)

# Display the encoded DataFrame
print(df_encoded)
```

```python
#   Discuss the need for converting categorical data into dummy variables (one-hot encoding).
# 	Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
# 	Display a portion of the transformed dataset

import pandas as pd
# Discuss the need for converting categorical data into dummy variables (one-hot encoding).
print("## One-Hot Encoding: Why It's Necessary")
print(
    "Many machine learning algorithms cannot directly work with categorical data. They require numerical input. One-hot encoding is a common technique to convert categorical data into a numerical format that algorithms can understand."
)
print(
    "In one-hot encoding, we create new binary (0 or 1) columns for each unique category within a categorical feature. Each row gets a 1 in the column corresponding to its category and 0s in all other columns. This avoids introducing an artificial order or magnitude to the categories."
)
print(
    "For example, if we have a 'Color' feature with categories 'Red', 'Green', and 'Blue', we'd create three new columns: 'Color_Red', 'Color_Green', and 'Color_Blue'. A row with 'Red' would have 'Color_Red' as 1 and the others as 0."
)

# Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
# (Assuming you have already identified the categorical columns in the DataFrame: categorical_cols)

# Import the necessary library
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder object
encoder = OneHotEncoder(handle_unknown='ignore')

# Fit and transform the categorical columns
encoded_data = encoder.fit_transform(df[categorical_cols])

# Convert the sparse matrix to a dense array
encoded_array = encoded_data.toarray()

# Create a new DataFrame with the encoded features
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_cols))

# Concatenate the encoded DataFrame with the original DataFrame
df_encoded = pd.concat([df, encoded_df], axis=1)

# Remove the original categorical columns if you don't need them anymore
df_encoded = df_encoded.drop(categorical_cols, axis=1)

# Display a portion of the transformed dataset
print("\n## Transformed Dataset (with One-Hot Encoding):")
print(df_encoded.head())  # Display the first few rows
```
